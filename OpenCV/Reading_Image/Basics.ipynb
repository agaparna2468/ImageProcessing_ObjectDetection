# Reading Images using OpenCV

retval = cv2.imread(filename[, flags])
//retval has the numpy array if image is successfully loaded otherwise None

filename can be relative or absolute

flags are used to read image in particular format. Values of flag:
default value of flags is 1 or cv2.IMREAD_COLOR which is used to load image as coloured image

0 or cv2.IMREAD_GRAYSCALE which is used to load image in grayscale

-1 or cv2.IMREAD_UNCHANGED which loads image as such with alpha channel(controls transparency of image).

```python
cb_img = cv2.imread("checkerboard_18x18.png", 0) #reads it as a numpy array in grayscale
#elements of array are intensity of pixel in unsigned 8 bits
print(cb_img.shape)
print(cb_img.dtype)
print(cb_img)
```

```python
plt.imshow(cb_img) #doesnot show black and white because colourmap used by imshow is different 
#it displays mathematical representation of image and not actual image so is not 18x18 pixels.
```

```python
plt.imshow(cb_img, cmap ='gray') #shows in colour map gray
#gray scale images have values between 0 and 255 where 255 shows white and 0 shows black
```

# Splitting and Merging Colour channels

cv2.split() => divides a multichannel array into several single channel arrays


cv2.merge() => merges several arrays to make a single multi-channeled array. All input matrices must be of same dimension

```python
img_bgr = cv2.imread("New_Zealand_Lake.jpg", 1) #OpenCv reads in bgr format

b, g, r = cv2.split(img_bgr) 
#b, g, r are 2D numpy array (matrices) which contain pixel intensities for each colour
```

```python
plt.figure(figsize = [20,5]) #width x height
plt.subplot(141); plt.imshow(r, cmap = 'gray'); plt.title("Red Channel")
plt.subplot(142); plt.imshow(g, cmap = 'gray'); plt.title("Green Channel")
plt.subplot(143); plt.imshow(b, cmap = 'gray'); plt.title("Blue Channel")

#represents each colour in gray scale with black showing absence of colour and white showing presence" 

imMerged = cv2.merge((b,g,r))
plt.subplot(144); plt.imshow(imMerged[:,:,::-1]); plt.title("Merged")
```

# Converting to different Color Space

We use function cv2.cvtColor(src, code) to convert between coloured spaces

dst = cv2.cvtColor(src, code) 
where src is source of image, code represents the type of conversion we want. Result will be a representation of image in different color space.

src input image could be 8 bit unsigned representation or 16 bit unsigned or single precision floating point

If conversion happens in RGB or BGR format, explicitily specify that. By default in OpenCV, BGR is used.

In standard 24 bit, first 1 byte is of blue then green then red

```python
#Changing from BGR to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  #This will have red as 1st channel.
#cv2.COLOR_BGR2RGB is a flag indicating to OpenCV that we are supplying BGR image and convert it to RGB
```

```python
plt.imshow(img_rgb)
```

# Changing to HSV Color Space 

HSV stands for hue, saturation and value

Hue stands for colour of image

Saturation represents intensity of colour that is if it is pure or dull

Value represents how light or dark the colour is

```python
Image(filename = "colour_spaces.jpg")
```

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
h, s, v = cv2.split(img_hsv)  #splitting and getting individual channels

plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(h, cmap = 'gray'); plt.title("Hue Channel");
plt.subplot(142); plt.imshow(s, cmap = 'gray'); plt.title("Saturation Channel");
plt.subplot(143); plt.imshow(v, cmap = 'gray'); plt.title("Value Channel");

plt.subplot(144); plt.imshow(img_rgb); plt.title("original")
```

```python
# Modifying Individual Channel

h_new = h + 10; #Shifting the position in colour position
img_merged2 = cv2.merge((h_new, s, v))

img_rgb2 = cv2.cvtColor(img_merged2, cv2.COLOR_HSV2RGB)

plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(img_rgb2); plt.title("New Merged");
plt.subplot(142); plt.imshow(img_rgb); plt.title("Original");
# Moved to blue
```

# Difference between Matplotlib imshow and Opnecv imshow

```python
cb_img = cv2.imread("checkerboard_18x18.png", 0)
window1 = cv2.namedWindow("w1") #create a named Window with name "w1"
cv2.imshow(window1, cb_img) # Call OpenCV version of imshow and pass in window object and image we want to show on that window\

cv2.waitKey(8000)
# Needs to be used with OpenCv imshow. 
# Argument is in milliseconds. Window is display for that long
# If we don't use this then imshow will display the window indefinitely and there will be no way to exit.
# window won't close until 8 secs

cv2.destroyWindow(window1)
```

```python
# On passing 0 to waitkey, image is displayed till any other key is pressed 
cv2.imshow(window1, cb_img) # Call OpenCV version of imshow and pass in window object and image we want to show on that window\

cv2.waitKey(0)
cv2.destroyWindow(window1)
```

```python
Alive = True

while(Alive):
    cv2.imshow(window1, cb_img)
    keypress = cv2.waitKey(1)    # returns the key pressed
    if keypress == ord('q'):
        Alive = False
        #exits only when 'q' is pressed
        
cv2.destroyAllWindows()
        
```
